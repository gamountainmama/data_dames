## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
ela21 = pd.read_csv("https://gamountainmama.github.io/data_dames/ELA21.csv")
ela22 = pd.read_csv("https://gamountainmama.github.io/data_dames/ELA22.csv")
math21 = pd.read_csv("https://gamountainmama.github.io/data_dames/MATH21.csv")
math22 = pd.read_csv("https://gamountainmama.github.io/data_dames/MATH22.csv")
sci22 = pd.read_csv("https://gamountainmama.github.io/data_dames/SCIE22.csv")
soc22 = pd.read_csv("https://gamountainmama.github.io/data_dames/SOCI22.csv")

ela21.head()
```

```python
df = ela21.merge(ela22, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df = df.merge(math21, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df = df.merge(math22, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df = df.merge(sci22, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df = df.merge(soc22, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df.head()
```

```python
# Drop rows that are missing data
df = df.dropna()
df.head()
```

```python
# Drop 'GTID', 'Course', and 'Scale Score'
df = df.drop(columns='GTID', axis=1)
df = df.drop(columns='Grade', axis=1)
df = df.drop(columns='MATH 2021 Pass', axis=1)
df = df.drop(columns='ELA 21-22 Level', axis=1)
df = df.drop(columns='ELA 20-21 Level', axis=1)
df = df.drop(columns='ELA 21-22 Pass', axis=1)
df = df.drop(columns='ELA 20-21 Pass', axis=1)
df = df.drop(columns='ELA 21-22 Scale Score', axis=1)
df = df.drop(columns='MATH 20-21 Level', axis=1)
df = df.drop(columns='MATH 21-22 Level', axis=1)
df = df.drop(columns='MATH 21-22 Scale Score', axis=1)
df = df.drop(columns='SCIE 21-22 Level', axis=1)
df = df.drop(columns='SCIE 21-22 Pass', axis=1)
df = df.drop(columns='SCIE 21-22 Scale Score', axis=1)
df = df.drop(columns='SOCI 21-22 Level', axis=1)
df = df.drop(columns='SOCI 21-22 Pass', axis=1)
df = df.drop(columns='SOCI 21-22 Scale Score', axis=1)

df.head()
```

```python
# Determine the number of unique values in each column.
df.dtypes
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
num_data = pd.get_dummies(df)
num_data.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = num_data.drop('MATH 21-22 Pass', axis=1).values
y = num_data['MATH 21-22 Pass'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input = X_train_scaled.shape[1]
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=20, activation="relu", input_dim=input))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=20, activation="tanh"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=20, verbose=1)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=1)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
