## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the combined_data2122.csv
import pandas as pd 
df = pd.read_csv("https://gamountainmama.github.io/data_dames/GeorgiaData_UnionCounty/combined_data2122.csv")
df.head()
```

```python
# Drop the non-beneficial ID column 'id'
df = df.drop(columns='id', axis=1)

df.head()
```

```python
# Determine the datatype of each column
df.dtypes
```

```python
# create a dataframe for 2022 test results
success_df = df[['Gender','Ethnicity','ELL','SWD','ED','SST','Gifted','Absences','Lexile','MATH 21 Scale Score','ELA 21 Scale Score','ELA 22 Pass','MATH 22 Pass','SCIE 22 Pass','SOCI 22 Pass']]
success_df.head()
```

```python
# add a column to determine how many subjects each student passed
success_df['Subjects Passed'] = success_df['ELA 22 Pass']+success_df['MATH 22 Pass']+success_df['SCIE 22 Pass']+success_df['SOCI 22 Pass']
success_df.head()
```

```python
# add a column to determine if each student passed overall (all four subjects)
success_df['Overall Pass'] = 0
success_df.loc[success_df['Subjects Passed'] == 4, 'Overall Pass'] = 1
success_df.head()
```

```python
# create a CSV of the new columns
success_df.to_csv('ga_data_overall_success.csv', index=False)
```

```python
from google.colab import files
files.download('ga_data_overall_success.csv')
```

```python
success_df.drop(['ELA 22 Pass','MATH 22 Pass','SCIE 22 Pass','SOCI 22 Pass','Subjects Passed'],inplace=True,axis=1)
success_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
success_num_data = pd.get_dummies(success_df)
success_num_data.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = success_num_data.drop('Overall Pass', axis=1).values
y = success_num_data['Overall Pass'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input = X_train_scaled.shape[1]
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=10, activation="relu", input_dim=input))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50, verbose=1)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=1)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
