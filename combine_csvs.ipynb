## Preprocessing

```python
# Import our dependencies
import pandas as pd

#  Import and read the CSV files.
ela21 = pd.read_csv("https://gamountainmama.github.io/data_dames/ELA21.csv")
ela22 = pd.read_csv("https://gamountainmama.github.io/data_dames/ELA22.csv")
math21 = pd.read_csv("https://gamountainmama.github.io/data_dames/MATH21.csv")
math22 = pd.read_csv("https://gamountainmama.github.io/data_dames/MATH22.csv")
sci22 = pd.read_csv("https://gamountainmama.github.io/data_dames/SCIE22.csv")
soc22 = pd.read_csv("https://gamountainmama.github.io/data_dames/SOCI22.csv")

ela21.head()
```

```python
df = ela21.merge(ela22, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df = df.merge(math21, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df = df.merge(math22, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df = df.merge(sci22, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df = df.merge(soc22, how = 'left', on = ['GTID', 'DaysAbsent', 'ELL', 'EconomicallyDisadvantaged', 'Ethnicity', 'GTID', 'Gender', 'Gifted', 'Grade', 'LastLexileScore', 'SST', 'SWD'])
df.head()
```

```python
# Drop rows that are missing data
df = df.dropna()
df.head()
```

```python
# Drop 'GTID' and 'Grade'
df = df.drop(columns='GTID', axis=1)
df = df.drop(columns='Grade', axis=1)
df.head()
```
