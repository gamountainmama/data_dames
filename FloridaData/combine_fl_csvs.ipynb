## Preprocessing

```python
# Import our dependencies
import pandas as pd

#  Import and read the CSV files.
arts = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/arts_data.csv")
comp = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/comp_sci_data.csv")
eng = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/english_data.csv")
entr = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/entrepreneurship_data.csv")
forlang = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/foreign_lang_data.csv")
hum = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/humanities_data.csv")
law = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/law_pol_data.csv")
math = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/math_data.csv")
psych = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/psychology_data.csv")
res = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/research_data.csv")
sci = pd.read_csv("https://gamountainmama.github.io/data_dames/FloridaData/science_data.csv")

comp['Course'].value_counts()
```

```python
eng = eng.drop(columns='Course', axis=1)
eng = eng.drop(columns='Course_Type', axis=1)
eng = eng.rename(columns={'Section_Grade': 'English Grade'})
eng.head()
```

```python
hum = hum.drop(columns='Course', axis=1)
hum = hum.drop(columns='Course_Type', axis=1)
hum = hum.rename(columns={'Section_Grade': 'Humanities Grade'})
hum.head()
```

```python
math = math.drop(columns='Course', axis=1)
math = math.drop(columns='Course_Type', axis=1)
math = math.rename(columns={'Section_Grade': 'Math Grade'})
math.head()
```

```python
sci = sci.drop(columns='Course', axis=1)
sci = sci.drop(columns='Course_Type', axis=1)
sci = sci.rename(columns={'Section_Grade': 'Science Grade'})
sci.head()
```

```python
comp = comp.drop(columns='Course', axis=1)
comp = comp.drop(columns='Course_Type', axis=1)
comp = comp.rename(columns={'Section_Grade': 'Computer Science Grade'})
comp.head()
```

```python
# join subject dataframes
df = eng.merge(math, how = 'left', on = ['StudentID', 'Grade'])
df = df.merge(sci, how = 'left', on = ['StudentID', 'Grade'])
df = df.merge(hum, how = 'left', on = ['StudentID', 'Grade'])
df = df.merge(comp, how = 'left', on = ['StudentID', 'Grade'])

#df = df.merge(entr, how = 'left', on = ['StudentID', 'Grade'])
#df = df.merge(forlang, how = 'left', on = ['StudentID', 'Grade'])
#df = df.merge(law, how = 'left', on = ['StudentID', 'Grade'])
#df = df.merge(math, how = 'left', on = ['StudentID', 'Grade'])
#df = df.merge(psych, how = 'left', on = ['StudentID', 'Grade'])
#df = df.merge(res, how = 'left', on = ['StudentID', 'Grade'])

df
```

```python
# Drop rows that are missing data
df = df.dropna()
df
```

```python
# Drop 'StudentID'
df = df.drop(columns='StudentID', axis=1)
df.head()
```

```python
# add a column to determine if each student made an A in Computer Science
df['Computer Science A'] = 0
df.loc[df['Computer Science Grade'] >= 90, 'Computer Science A'] = 1
df.head()
```
